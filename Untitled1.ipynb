```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Parmetros do filtro
frequencia_corte = 5.0  # Frequncia de corte em Hz
frequencia_amostragem = 50.0  # Frequncia de amostragem em Hz
ordem = 4  # Ordem do filtro

# Funo para criar um filtro Butterworth
def filtro_passa_baixa(cutoff, fs, ordem):
    nyquist = 0.5 * fs  # Frequncia de Nyquist
    normal_cutoff = cutoff / nyquist  # Normalizao da frequncia
    b, a = butter(ordem, normal_cutoff, btype='low', analog=False)
    return b, a

# Funo para aplicar o filtro nos dados
def aplicar_filtro(dados, b, a):
    return lfilter(b, a, dados)

# Gerao de um sinal ruidoso (seno + rudo)
tempo = np.linspace(0, 5, 5 * int(frequencia_amostragem), endpoint=False)
sinal = np.sin(2 * np.pi * 1.2 * tempo) + 0.5 * np.random.randn(len(tempo))

# Aplicao do filtro passa-baixa
b, a = filtro_passa_baixa(frequencia_corte, frequencia_amostragem, ordem)
sinal_filtrado = aplicar_filtro(sinal, b, a)

# Plotando o sinal original e o filtrado
plt.figure(figsize=(10, 5))
plt.plot(tempo, sinal, label='Sinal Original')
plt.plot(tempo, sinal_filtrado, label='Sinal Filtrado', linewidth=2)
plt.legend()
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('Filtro Passa-Baixa em Python')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Parmetros do filtro
frequencia_corte = 5.0  # Frequncia de corte em Hz
frequencia_amostragem = 50.0  # Frequncia de amostragem em Hz
ordem = 4  # Ordem do filtro

# Funo para criar um filtro Butterworth
def filtro_passa_baixa(cutoff, fs, ordem):
    nyquist = 0.5 * fs  # Frequncia de Nyquist
    normal_cutoff = cutoff / nyquist  # Normalizao da frequncia
    b, a = butter(ordem, normal_cutoff, btype='low', analog=False)
    return b, a

# Funo para aplicar o filtro nos dados
def aplicar_filtro(dados, b, a):
    return lfilter(b, a, dados)

# Gerao de um sinal ruidoso (seno + rudo)
tempo = np.linspace(0, 5, 5 * int(frequencia_amostragem), endpoint=False)
sinal = np.sin(2 * np.pi * 1.2 * tempo) + 0.5 * np.random.randn(len(tempo))

# Aplicao do filtro passa-baixa
b, a = filtro_passa_baixa(frequencia_corte, frequencia_amostragem, ordem)
sinal_filtrado = aplicar_filtro(sinal, b, a)

# Plotando os sinais discretizados
plt.figure(figsize=(12, 6))

# Sinal Original Discretizado
plt.subplot(2, 1, 1)
plt.stem(tempo, sinal, use_line_collection=True, label='Sinal Original (Discretizado)')
plt.legend()
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('Sinal Original Discretizado')

# Sinal Filtrado Discretizado
plt.subplot(2, 1, 2)
plt.stem(tempo, sinal_filtrado, use_line_collection=True, linefmt='r-', markerfmt='ro', label='Sinal Filtrado (Discretizado)')
plt.legend()
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('Sinal Filtrado (Passa-Baixa) Discretizado')

plt.tight_layout()
plt.show()
```
